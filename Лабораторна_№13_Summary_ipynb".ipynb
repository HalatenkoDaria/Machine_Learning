```python
import spacy
from collections import Counter
import heapq
import re
```

```python
# Завантаження моделі
nlp = spacy.load("en_core_web_sm")
```

```python
# Вхідний текст (приклад)
text = """
SMS spam detection is an important task in modern NLP applications.
Users receive many unwanted messages daily, and it's crucial to filter them.
Using models like Logistic Regression or Naive Bayes helps classify messages effectively.
Balancing datasets with SMOTE can improve accuracy for rare spam messages.
TF-IDF is a common vectorization method that enhances word relevance in classification.
"""
```

```python
# Попередня обробка
text = re.sub(r'\s+', ' ', text)
doc = nlp(text)

# Отримуємо речення
sentences = list(doc.sents)

# Створюємо словник важливості слів
word_frequencies = {}
for token in doc:
    if token.is_alpha and not token.is_stop:
        word_frequencies[token.text] = word_frequencies.get(token.text, 0) + 1

# Нормалізуємо частоти
max_freq = max(word_frequencies.values())
for word in word_frequencies:
    word_frequencies[word] /= max_freq

# Оцінюємо речення
sentence_scores = {}
for sent in sentences:
    for token in sent:
        if token.text in word_frequencies:
            sentence_scores[sent] = sentence_scores.get(sent, 0) + word_frequencies[token.text]

# Вибір 3 найсильніших речень
summary_sentences = heapq.nlargest(3, sentence_scores, key=sentence_scores.get)

# Формуємо резюме
summary = " ".join([sent.text for sent in summary_sentences])
print("🔍 Summary:\n", summary)
```

Використання трансформерів для створення summary

```python
from transformers import pipeline

# Створюємо пайплайн для summarization
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Вхідний текст
text = """
SMS spam detection is an important task in modern NLP applications.
Users receive many unwanted messages daily, and it's crucial to filter them.
Using models like Logistic Regression or Naive Bayes helps classify messages effectively.
Balancing datasets with SMOTE can improve accuracy for rare spam messages.
TF-IDF is a common vectorization method that enhances word relevance in classification.
"""

# Генерація summary
summary = summarizer(text, max_length=60, min_length=20, do_sample=False)

# Вивід
print("📌 Summary:")
print(summary[0]['summary_text'])
```

Цей метод формує Summary  своїми словами, а не просто вибирає речення. Працює без додаткової очистки — достатньо тексту.

**Висновок:** У цьому дослідженні ми протестували два підходи до автоматичного створення стислого змісту тексту. Перший метод — з використанням бібліотеки spaCy — базується на підрахунку частот слів і виборі найбільш інформативних речень. Другий метод — із використанням бібліотеки Hugging Face Transformers — формує резюме «своїми словами» за допомогою нейронної мережі. Обидва підходи дали осмислені результати, але трансформер продемонстрував більшу гнучкість та якість узагальнення без необхідності складної попередньої обробки.
